```python


# Clean up course names
df['Course'] = df['Course'].str.upper().str.strip()

# Create a combined column for x-axis
df['Year_Course'] = df['Year'].astype(str) + ' - ' + df['Course']

# Sort the order of combined labels
x_order = sorted(df['Year_Course'].unique())

# Custom color palette
custom_palette = {
    'Yes': 'green',
    'No': 'red',
    'Looking': 'yellow'
}

# Set figure size
plt.figure(figsize=(12, 6))

# Plot
sns.countplot(
    data=df,
    x='Year_Course',
    hue='Internship_Done',
    hue_order=['Yes', 'No', 'Looking'],
    palette=custom_palette,
    order=x_order
)

# Labels and titles
plt.title('Internship Status by Year and Course')
plt.xlabel('Year - Course')
plt.ylabel('Number of Students')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Internship Status')
plt.tight_layout()
plt.show()
```

```python

# Step 2: Clean the 'Course' column - convert to uppercase and remove extra spaces
df['Course'] = df['Course'].astype(str).str.upper().str.strip()

# Step 3: Clean the 'Year' column - just in case
df['Year'] = df['Year'].astype(str).str.title().str.strip()

# Step 4: Filter data only for MCA course
mca_df = df[df['Course'] == 'MCA']

# Step 5: Filter only specific years for MCA: 1st, Final Year, Passed Out
allowed_years = ['1st', 'Final Year', 'Passed Out']
mca_df = mca_df[mca_df['Year'].isin(allowed_years)]

# Step 6: Plot using seaborn countplot
sns.set(style="whitegrid")

# Define custom colors: 'Yes' = Green, 'No' = Red, 'Looking' = Yellow
custom_colors = {"Yes": "green", "No": "red", "Looking": "yellow"}

plt.figure(figsize=(8, 5))
sns.countplot(
    data=mca_df,
    x="Year",
    hue="Internship_Done",
    palette=custom_colors,
    order=allowed_years,
    hue_order=["Yes", "No", "Looking"]
)

# Step 7: Add labels and title
plt.title("MCA Students Internship Status")
plt.xlabel("Year")
plt.ylabel("Number of Students")
plt.legend(title="Internship Done")
plt.tight_layout()
plt.show()
```
