```python

df['Course'] = df['Course'].str.strip().str.upper()  # Normalize case and trim
# Convert columns to numeric
df['Intern_Learn_Score'] = pd.to_numeric(df['Intern_Learn_Score'], errors='coerce')
df['Placement_Score'] = pd.to_numeric(df['Placement_Score'], errors='coerce')

# Drop rows with both scores missing
df = df.dropna(subset=['Intern_Learn_Score', 'Placement_Score'], how='all')

# Group by Course (average scores for each course)
grouped = df.groupby('Course')[['Intern_Learn_Score', 'Placement_Score']].mean().reset_index()

# Bar chart positions
x = np.arange(len(grouped['Course']))  # number of courses
width = 0.35  # bar width

# Plot
grouped.plot(x='Course', y=['Intern_Learn_Score', 'Placement_Score'], kind='line', marker='o')
plt.title('Intern Learn vs Placement Score by Course')
plt.xlabel('Course')
plt.ylabel('Average Score')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python

# Clean and normalize course names
df['Course'] = df['Course'].astype(str).str.strip().str.upper()

# Convert columns to numeric (non-convertible become NaN)
df['Intern_Learn_Score'] = pd.to_numeric(df['Intern_Learn_Score'], errors='coerce')
df['Placement_Score'] = pd.to_numeric(df['Placement_Score'], errors='coerce')

# Drop rows where 'Course' is NaN (to avoid plotting issues)
df = df[df['Course'].notna()]

# Instead of dropping rows where both scores are missing, we keep all rows.
# Group by course and take mean, NaNs will be ignored in mean()
grouped = df.groupby('Course')[['Intern_Learn_Score', 'Placement_Score']].mean().reset_index()

# Print to confirm MBA is included
print("Grouped Data:\n", grouped)

# Plot line chart with markers
plt.figure(figsize=(10, 6))
plt.plot(grouped['Course'], grouped['Intern_Learn_Score'], marker='o', label='Intern Learn Score', color='skyblue')
plt.plot(grouped['Course'], grouped['Placement_Score'], marker='o', label='Placement Score', color='salmon')

# Labels and formatting
plt.title('Intern Learn vs Placement Score by Course')
plt.xlabel('Course')
plt.ylabel('Average Score')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```
