```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
import seaborn as sns
 
```

```python
df = pd.read_csv(r"C:\Users\laxmi\Documents\JUPYTER\PROJECT\Placement.csv")
```

```python
df.replace(['Na', 'Nan', 'N', 'NaN', np.nan], '', inplace=True)
```

```python
df.fillna('')
```

```python
df
```

```python

# Step 2: Clean the 'Course' column - convert to uppercase and remove extra spaces
df['Course'] = df['Course'].astype(str).str.upper().str.strip()

# Step 3: Clean the 'Year' column - just in case
df['Year'] = df['Year'].astype(str).str.title().str.strip()

# Step 4: Filter data only for MCA course
mca_df = df[df['Course'] == 'MCA']

# Step 5: Filter only specific years for MCA: 1st, Final Year, Passed Out
allowed_years = ['1st', 'Final Year', 'Passed Out']
mca_df = mca_df[mca_df['Year'].isin(allowed_years)]

# Step 6: Plot using seaborn countplot
sns.set(style="whitegrid")

# Define custom colors: 'Yes' = Green, 'No' = Red, 'Looking' = Yellow
custom_colors = {"Yes": "#2ca02c", "No": "red", "Looking": "yellow"}

plt.figure(figsize=(8, 5))
sns.countplot(
    data=mca_df,
    x="Year",
    hue="Placement_Attempted",
    palette=custom_colors,
    order=allowed_years,
    hue_order=["Yes", "No", "Looking"]
)

# Step 7: Add labels and title
plt.title("MCA Students Placement Status")
plt.xlabel("Year")
plt.ylabel("Number of Students")
plt.legend(title="Placement Done")
plt.tight_layout()
plt.show()
```

```python


# Normalize course names
# df['Course'] = df['Course'].str.strip().str.upper()

# Convert score columns to numeric
df['Placement_Score'] = pd.to_numeric(df['Placement_Score'], errors='coerce')
# df['Intern_Learn_Score'] = pd.to_numeric(df['Intern_Learn_Score'], errors='coerce')

# Drop rows where both scores are missing
df = df.dropna(subset=['Placement_Score'], how='all')

# Group by Course and take mean
grouped = df.groupby('Course')[ 'Placement_Score'].mean().reset_index()

# Plot Placement Score by Course
plt.figure(figsize=(10, 6))
plt.plot(grouped['Course'], grouped['Placement_Score'], marker='o', label='Placement Score', color='salmon')

plt.title('Placement Score by Course')
plt.xlabel('Course')
plt.ylabel('Average Score')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

```python

# df["Intern_Learn_Label"].value_counts().plot(kind='bar', color='lightgreen')
# plt.title("Internship Learning Experience Ratings")
# plt.xlabel("Rating")
# plt.ylabel("Number of Students")
# plt.xticks(rotation=0)
# plt.show()


# Step 1: Normalize the labels
df["Placement _rating_Label"] = df["Placement _rating_Label"].str.title()

# Step 2: Define the color mapping
color_map = {
    'Excellent': 'Orange',
    'Good': 'lightgreen',
    'Average': '#1f77b4',
    'Below Average': 'yellow',
    'Poor': 'red'
}

# Step 3: Filter value counts to include only known labels
value_counts = df["Placement _rating_Label"].value_counts()
value_counts = value_counts.loc[value_counts.index.isin(color_map.keys())]

# Step 4: Match colors to labels
colors = [color_map[label] for label in value_counts.index]

# Step 5: Plot
value_counts.plot(kind='bar', color=colors)
plt.title("Placement Experience Ratings")
plt.xlabel("Rating")
plt.ylabel("Number of Students")
plt.xticks(rotation=0)
plt.show()



```
